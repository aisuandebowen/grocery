```python
import torch
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
tensor = torch.tensor(arr)

tensor
```

```python
# 0~1 正态分布 第一位参数是可变参数 可 int、list、tuple
tensor1 = torch.randn([2,2])
tensor1
```

```python
# 0~1 均匀分布
tensor2 = torch.rand((1,2))
tensor2
```

```python
tensor3 = torch.normal(1,2,(2,3))
tensor3
```

```python
tensor4 = torch.normal(1,torch.arange(1,5,dtype=torch.float))
tensor4
```

```python
tensor5 = torch.normal(torch.arange(1,5,dtype=torch.float),2)
tensor5
```

```python
# 自己设置step
t6 = torch.arange(torch.tensor(1),5,1, dtype=torch.float)
# 等间距
t7 = torch.linspace(1,5,4,dtype=torch.float)

print(t6,t7)
```

```python
torch.manual_seed(1)
tt1 = torch.rand(1,2)
print(tt1)
torch.manual_seed(1)
tt2 = torch.rand(1,2)
print(tt2)
```
